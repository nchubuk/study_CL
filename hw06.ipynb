С помощью os.listdir() обходим все файлы с музыкой в папке; при этом в папке, которую мы обходим, нужные нам файлы называются в таком формате:

    имя исполнителя - номер песни - название песни.mp3
    
(при этом дефисы могут попасться как в имени исполнителя, так и в названии песни тоже). Программа должна:

1. Обрабатывать только файлы, которые удовлетворяют этому шаблону (открывать их не будем, давайте будем просто сообщать, что X - это песенка исполнителя Y)
2. По итогам работы собрать статистику с перечнем исполнителей и их песен в виде:

        имя исполнителя:
            номер песни. название песни
        имя исполнителя:
            номер песни. название песни

```python
import os
from collections import defaultdict
```

```python
tracks = [re.findall(r'(.+) ?- ?(\d+) ?- ?(.+)(?:\.mp3)', f) for f in os.listdir('/content/music')]
tracks_dict = defaultdict(list)

for track in tracks:
  if track != []:
    tracks_dict[track[0][0]].append(f"{int(track[0][1])}. {track[0][2]}")

for key, value in sorted(tracks_dict.items()):
  print(f'{key}:', *value, sep='\n\t')
```

#### Задача 13.

Увеличить все встретившиеся в тексте четные числа на 1, а нечетные на 2, в том числе и отрицательные. Перед числом может стоять знак. Считаем, что между знаком и числом пробела нет.

```python
text13 = '''
21 марта 1855 года. Ничего не делал. Получил восхитительное письмо от Маши, в котором она описывает мне свое знакомство с Тургеневым. Милое, славное письмо, возвысившее меня в собственном мнении и побуждающее к деятельности
4 июня 1856 года. Встал в 5, гулял, признаюсь, с ужасно эротическими мыслями. Читал первые стихотворения Пушкина. Потом разбирал свои старые тетради, непонятная, но милая дичь
'''
list_txt13 = list(text13)
for i in re.finditer(r'\d+', text13):
  if int(i.group()) % 2:
    list_txt13[i.start():i.end()] = str(int(i.group()) + 2)
  else:
    list_txt13[i.start():i.end()] = str(int(i.group()) + 1)

print(''.join(list_txt13))
```
