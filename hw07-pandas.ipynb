#### Задание 1.

Соберите всю информацию о студентах в одну таблицу df. В получившейся таблице должна быть информация и оценки всех студентов из всех групп. Напечатайте несколько строк таблицы для демонстрации результата.¶

```python
import os
import re

def GetFrame(rpath):
  path = os.path.normpath(rpath)
  students_all = pd.DataFrame()
  marks_all = pd.DataFrame()
  for f in os.listdir(path):
    if re.match(r'Students_info_\d\.csv', f):
      tmp = pd.read_csv(os.path.join(path, f))
      students_all = pd.concat([students_all, tmp])
    elif re.match(r'Students_marks_\d\.csv', f):
      tmp = pd.read_csv(os.path.join(path, f))
      marks_all = pd.concat([marks_all, tmp])
  return pd.merge(students_all, marks_all, on='index') # при пропуске одного из изначальных файлов по студентам или оценкам - в объединение они не попадут
```

#### Задание 14.

Сколько студентов успешно сдали все экзамены?

Создайте столбец OverAll_PassStatus и запишите в него для каждого студента 'F', если студент не сдал хотя бы один из трех экзаменов, а иначе 'P'.

Посчитайте количество студентов, которые сдали все экзамены.

```python
import numpy as np

st_all['OverAll_PassStatus'] = np.where(
    (st_all['Math_PassStatus'] == 'F') |
    (st_all['Reading_PassStatus'] == 'F') |
    (st_all['Writing_PassStatus'] == 'F'),
    'F',
    'P'
)
```

```python
print(f"Кол-во студентов, которые сдали все экзамены: {st_all['OverAll_PassStatus'].value_counts().get('P', 0)}")
```

#### Задание 15.

Переведем баллы в оценки

*Система перевода баллов в оценки*

        [90;100] = A
        [80;90) = B
        [70;80) = C
        [60;70) = D
        [50;60) = E
        [0;50) = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

```python
def GetGrade(average_mark):
  if 90 <= average_mark <= 100:
    return 'A'
  elif 80 <= average_mark < 90:
    return 'B'
  elif 70 <= average_mark < 80:
    return 'C'
  elif 60 <= average_mark < 70:
    return 'D'
  elif 50 <= average_mark < 60:
    return 'E'
  else:
    return 'F'

st_all['Grade'] = st_all[['math score', 'reading score', 'writing score']].mean(axis=1).apply(GetGrade)
```

```python
st_all['Grade'].value_counts()
```
