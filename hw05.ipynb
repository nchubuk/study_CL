Реализуйте с помощью классов базу данных для ветеринарного магазина из предыдущего семинара. Видимо, нам понадобятся два класса: для владельцев животных и для самих животных. У владельца мы храним имя, фамилию и номер телефона. У животного пусть будет кличка, возраст и дополнительный атрибут "записан на прием", в котором будет храниться информация о дате следующего приема (если ее нет - None). База данных должна храниться в бинарном файле (используйте pickle или dill). При запуске программы должны быть доступны несколько действий: завести нового клиента, завести новое животное для существующего клиента и записать животное на прием (или отметить, что прием прошел и новой записи нет). Сами экземпляры клиентов хранятся просто в списке.

Примечание: в атрибутах могут храниться не только строки, но и списки, но не забывайте об их особенностях.

```python
from datetime import date

class Owner:
  def __init__(self, name, surname, phone):
    self.name = name
    self.surname = surname
    self.phone = phone
    self.pets = []

  def __str__(self):
    return f'{self.name} {self.surname} {self.phone} {self.pets}'

  def __repr__(self):
    return f"Владелец('{self.name}', '{self.surname}', '{self.phone}', '{self.pets}')"

  def add_pet(self, Pet):
    self.pets.append(Pet)

class Pet:
  def __init__(self, name, age, appointment=None):
    self.name = name
    self.age = age
    self.appointment = appointment

  def __str__(self):
    return f'{self.name} {self.age} {self.appointment}'

  def __repr__(self):
    return f"Домашнее животное('{self.name}', '{self.age}', '{self.appointment}')"

  def book_appointment(self, appointment_date):
    if self.appointment == None:
      self.appointment = appointment_date
      print(f"Запись на прием успешно добавлена, дата встречи - {self.appointment}")
    elif self.appointment == date.today():
      print(f"Дата приема уже назначена на сегодня")
    elif self.appointment > date.today():
      self.appointment = appointment_date
      print(f"Дата записи приема ветеринаром обновлена на {self.appointment}")
    else:
      self.appointment = appointment_date
      print(f"Дата новой записи на прием - {self.appointment}")

  def cancel_appointment(self):
    if self.appointment == None:
      print("Записей о приеме нет")
    elif self.appointment < date.today():
      self.appointment = None
      print("Дата приема прошла")
    else:
      self.appointment = None
      print("Запись отменена")
```

```python
maxim = Owner('Максим', 'Бойко', '+79908887777')
bobik = Pet('Бобик', 5)
```
